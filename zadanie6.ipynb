```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split

sns.set(font_scale=1.3)
np.random.seed(42)

from sklearn.datasets import load_iris
raw_data=load_iris()
all_data=raw_data.copy()
data=all_data['data']
target=all_data['target']
feature_names=all_data['feature_names']
target_names=all_data['target_names']

df=pd.DataFrame(data=np.c_[data,target],columns=feature_names+['target'])
#podzial dane wejsciowe pierwsze 2 kolumny z data
#kolumna target jako int
x = data[:,[0,1]]
y = df['target'].apply(int).values

print(f'{x[:5]}\n')
print(f'{y[:5]}')
```

```
#zadanie 2 podzielenie danych na trenujace i testujace 
X_train, X_test, y_train, y_test = train_test_split(x,y)
```

```
from sklearn.svm import SVC
from sklearn import svm, datasets 
from mlxtend.plotting import plot_decision_regions
from sklearn import svm

#jadro rbf
svc = svm.SVC(kernel ='rbf', C = 5000).fit(x, y) 
plot_decision_regions(x, y, clf=svc, legend=4)

```

```
from sklearn.svm import SVC
from sklearn import svm, datasets 
from mlxtend.plotting import plot_decision_regions
from sklearn import svm

#jadro linear

svc = svm.SVC(kernel ='linear', C = 1).fit(x, y) 
plot_decision_regions(x, y, clf=svc, legend=2)

```
