```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs

data, _=make_blobs(n_samples=1000,centers=2,random_state=42)

df=pd.DataFrame(data,columns=['x1','x2'])
px.scatter(df,'x1','x2')
```

```
from sklearn.cluster import DBSCAN
cluster =DBSCAN(eps=0.3,min_samples=5)
cluster.fit(data)
```

```
df['cluster']=cluster.labels_
px.scatter(df, 'x1','x2','cluster',color_continuous_midpoint=0)
```

```
df['cluster']=cluster.labels_
px.scatter(df, 'x1','x2','cluster',template="plotly_dark",color_continuous_midpoint=0)
```
